## Feature Generation

```python
def getSubjectivity(text):
  return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
  return TextBlob(text).sentiment.polarity

def analyze(score, task="polarity"):
  if task=="subjectivity":
    if score < 1/3:
      return "low"
    elif score > 1/3:
      return "high"
    else:
      return "medium"
  else:
    if score < 0:
      return 'Negative'
    elif score == 0:
      return 'Neutral'
    else:
      return 'Positive'

def sentiment_analysis(df, text_col):

  df['subjectivity'] = df[text_col].apply(getSubjectivity).apply(lambda x: analyze(x, "subjectivity"))
  df['polarity'] = df[text_col].apply(getPolarity).apply(analyze)
  return df
```

```python
sentiments = sentiment_analysis(songDf, "track_name")
sentiments.head()
```

```python
def ohe(df, column, prefix):

  ohe_feature = pd.get_dummies(df[column])
  feature_name = ohe_feature.columns
  ohe_feature.columns = [prefix + "_" + str(i) for i in feature_name]
  ohe_feature.reset_index(drop = True, inplace = True)
  return ohe_feature
```

```python
def create_feature(df, cols):

  tfidf = TfidfVectorizer()
  matrix = tfidf.fit_transform(songDf['genres_list'].apply(lambda x: " ".join(x)))
  genre_df = pd.DataFrame(matrix.toarray())
  genre_df.columns = ['genre' + "_" + i for i in tfidf.get_feature_names_out()]
  genre_df.drop(columns = 'genre_unknown')
  genre_df.reset_index(drop = True, inplace = True)

  df = sentiment_analysis(df, "track_name")

  subject_ohe = ohe(df, 'subjectivity', 'subject') * 0.3
  polar_ohe = ohe(df, 'polarity', 'polar') * 0.5
  key_ohe = ohe(df, 'key', 'key') * 0.5
  mode_ohe = ohe(df, 'mode', 'mode') * 0.5

  popularity = df[["artist_pop","track_pop"]].reset_index(drop = True)
  scaler = MinMaxScaler()
  popularity_scaled = pd.DataFrame(scaler.fit_transform(popularity), columns = popularity.columns) * 0.2

  floats = df[cols].reset_index(drop = True)
  scaler = MinMaxScaler()
  floats_scaled = pd.DataFrame(scaler.fit_transform(floats), columns = floats.columns) * 0.2

  final = pd.concat([genre_df, floats_scaled, popularity_scaled, subject_ohe, polar_ohe, key_ohe, mode_ohe], axis = 1)

  final['id'] = df['id'].values

  return final
```

```python
def generate_playlist(feature_set, df):

  feature_set_playlist = feature_set[feature_set['id'].isin(df['id'].values)]

  feature_set_nonplaylist = feature_set[~feature_set['id'].isin(df['id'].values)]
  feature_set_final = feature_set_playlist.drop(columns = "id")
  return feature_set_final.sum(axis = 0), feature_set_nonplaylist
```

```python
def recommend(df, features, nonplaylist_feature):

  nonplaylist_df = df[df['id'].isin(nonplaylist_feature['id'].values)]

  nonplaylist_df['sim'] = cosine_similarity(nonplaylist_feature.drop('id', axis = 1).values, features.values.reshape(1, -1))[:,0]
  non_playlist_top40 = nonplaylist_df.sort_values('sim', ascending = False).head(40)

  return non_playlist_top40
```

```python
song_recommend = recommend(songDf, feature_set_vector, feature_set_nonplaylist)
song_recommend.head(10) 
```
