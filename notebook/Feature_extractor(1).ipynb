```python
from script.uri import uri_to_feature
import pandas as pd
from tqdm import tqdm
import re
```

```python
#reading the raw data file
path = '../data/raw_data.csv'
df = pd.read_csv(path)
df.head()
```

```python
# converting track uri into more 'readable' form
df["track_uri"] = df["track_uri"].apply(lambda x: re.findall(r'\w+$',x)[0])
df["track_uri"]
```

```python
testDf = df
```

# FEATURES

```python
uri_to_feature(df["track_uri"][0])
```

# EXCTRACTION
  Extracting above features of songs using uri from raw data and spotipy


```python
first_part = df["track_uri"].unique()[:10000]
second_part = df["track_uri"].unique()[10000:20000]
third_part = df["track_uri"].unique()[20000:]
datalist = [first_part, second_part, third_part]
```

```python
featurelist = []

for i in tqdm([uri for uri in datalist[0]]):
  try:
    featurelist.append(uri_to_feature(i))
  except:
    continue
```

```python
for i in tqdm([uri for uri in datalist[1]]):
  try:
    featurelist.append(uri_to_feature(i))
  except:
    continue
```

```python
for i in tqdm([uri for uri in datalist[2]]):
  try:
    featurelist.append(uri_to_feature(i))
  except:
    continue
```

# Finishing and Exporting

```python
featureDf = pd.DataFrame(featurelist)
new_df = pd.merge(featureDf, testDf, left_on = "track_uri", right_on="id")
new_df
```

```python
new_df.to_csv("../data/processed_data.csv")
```
